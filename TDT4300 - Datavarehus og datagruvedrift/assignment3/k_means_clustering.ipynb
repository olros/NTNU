## Function

A function for plotting that we are going to use later on.

````python
def plot_clusters(data, centroids):
    """
    Shows a scatter plot with the data points clustered according to the centroids.
    """
    # Assigning the data points to clusters/centroids.
    clusters = [[] for _ in range(centroids.shape[0])]
    for i in range(data.shape[0]):
        distances = np.linalg.norm(data[i] - centroids, axis=1)
        clusters[np.argmin(distances)].append(data[i])

    # Plotting clusters and centroids.
    fig, ax = plt.subplots()
    for c in range(centroids.shape[0]):
        if len(clusters[c]) > 0:
            cluster = np.array(clusters[c])
            ax.scatter(cluster[:, 0], cluster[:, 1], s=7)
    ax.scatter(centroids[:, 0], centroids[:, 1], marker='x', s=200, c='red')
````

## Implementation

A human can with an ease find five distinct clusters just by watching the plot. A computer, however, needs to be told how to find the clusters.

**Exercise:**

Implement the k-means clustering algorithm.

* Use the Euclidean (L<sub>2</sub>) distance.
* It is sufficient to use the basic Python constructs in your implementation, even though we heavily rely on Numpy throughout this assignment.

````python
def euclidiean_distance(x,y):
    return np.sum(np.sqrt(np.square(x - y)), axis=1)


def calculate_euclidean_distance(centroids,data, init_array):
    for i, centroid in enumerate(centroids):
        init_array[:, i] = euclidiean_distance(data, centroid)
    return init_array

def calculate_closet_cluster(distance_to_centroids):
    closest_clusters = np.zeros_like(distance_to_centroids) 
    for i, distances in enumerate(distance_to_centroids):
        min_index = np.argmin(distances)
        closest_clusters[i][min_index] = 1
    return closest_clusters

def kmeans(data, centroids):
    """
    Function implementing the k-means clustering.
    
    :param data
        data
    :param centroids
        initial centroids
    :return
        final centroids
    """
    ### START CODE HERE ###    
    shape = (data.shape[0], centroids.shape[0])
    distance_to_centroids = np.zeros(shape)
    current_centroids = centroids.copy()
    next_centroids = centroids.copy()
    running = True

    while running:
        next_centroids = current_centroids.copy()

        
        distance_to_centroids = calculate_euclidean_distance(next_centroids, data, distance_to_centroids)

        closest_clusters =  calculate_closet_cluster(distance_to_centroids)

        for i in range(len(next_centroids)):
            next_centroids[i] = np.mean(data[closest_clusters[:,i] == 1], axis=0)
        is_unchanged = np.array_equal(current_centroids, next_centroids) 
        if is_unchanged:
            centroids = next_centroids
            running = False
        
        current_centroids = next_centroids.copy()
    ### END CODE HERE ###
    return centroids
````

## Evaluation of Clustering

Silhouette Coefficient is an example of a measure for validation of the cluster quality.

**Exercise:**

Implement a function calculating the mean Silhouette Coefficient of all samples.

* Use the Euclidean (L<sub>2</sub>) distance.
* It is sufficient to use the basic Python constructs in your implementation, even though we heavily rely on Numpy throughout this assignment.

````python
def avarage_distance(x, y):
    return np.mean(np.sqrt(np.sum(np.square(x - y), axis=1)))

def create_a(current_cluster):
    a = np.zeros(len(current_cluster))
    for point in range(len(current_cluster)):
        a[point] = avarage_distance(current_cluster[point], current_cluster)
    return a 

def create_b(current_cluster, centroids, centroid, data, closest_clusters):
        b = np.zeros(len(current_cluster))
        for point in range(len(current_cluster)):
            lowest_dist = np.Inf
            for current_centroids_index, other_centroids in enumerate(centroids):
                if not np.array_equal(other_centroids, centroid):
                    cluster = data[closest_clusters[:, current_centroids_index] == 1]
                    cluster_dist = avarage_distance(current_cluster[point], cluster)
                    if lowest_dist > cluster_dist:
                        b[point] = cluster_dist
                        lowest_dist = cluster_dist
        return b
    

def silhouette_score(data, centroids):
    """
    Function implementing the k-means clustering.
    
    :param data
        data
    :param centroids
        centroids
    :return
        mean Silhouette Coefficient of all samples
    """
    ### START CODE HERE ### 
    shape = (data.shape[0], centroids.shape[0])
    scores = np.zeros(len(data))
    distance_to_centroids = np.zeros(shape)
    distance_to_centroids = calculate_euclidean_distance(centroids, data, distance_to_centroids)
    closest_clusters =  calculate_closet_cluster(distance_to_centroids)
    for i, centroid in enumerate(centroids):
        current_cluster = data[closest_clusters[:,i] == 1] 
        
        a = create_a(current_cluster)
        b = create_b(current_cluster, centroids, centroid, data, closest_clusters)
        
        
        scores[closest_clusters[:, i] == 1] = (b - a)/np.maximum(a, b)
    score = np.mean(scores)
    ### END CODE HERE ### 
    return score

    
````
